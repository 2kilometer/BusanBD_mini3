### 1. roop1에서 사용할 함수

```python
### 일상 대화 규칙을 딕셔너리{index:[단어1, 단어2]}로 반환
def getRulesDict1():
    ## 규칙이 담길 딕셔너리
    chat_dict = {}
    ## 단어별 번호(0~)
    row = 0
    ## 규칙을 딕셔너리에 담기
    # - chatbot_data['rule']에 있는 데이터를 불러와 '|'를 기준으로 자르고, 숫자를 key로, 단어를 value로 chat_dict에 저장.
    for rule in chatbot_data['rule']:
        chat_dict[row] = rule.split('|')  # row(0~숫자) 별로 입력된 단어를 '|'기준으로 잘라 chat_dict에 추가
        row += 1  # row에 1씩 더하기
    return chat_dict
getRulesDict1()
```

```python
### getRulesDict1에서 받은 딕셔너리를 받아 응답하는 함수
def chat1(request) :
    chat_dict = getRulesDict1()
    for k, v in chat_dict.items() :
        index = -1
        for word in v :
            try :
                if index == -1 :
                    index = request.index(word)
                else :
                    if index < request.index(word, index) :
                        index = request.index(word, index)
                    else :
                        index = -1
                        break
            except ValueError : 
                index = -1
                break
        if index > -1 :
            return chatbot_data['response'][k]
    return '무슨 말인지 모르겠어요.'
chat1("느그아부지뭐하시노")
```

```python
def getJinguMap():
    import folium
    ### 지도시각화 정의
    rv_map = folium.Map(location=[35.1665, 129.0431], tiles="CartoDB positron", zoom_start=13)
    layer_control = folium.LayerControl()
    
    ### 값 필터 설정
    # - 동 리스트 고윳값 정의
    dong_list = df_map["dong"].unique()
    # - 마커 색상 리스트
    color_list = ["#e51111", "#6c78d7", "#403157",
                 'b43838','#4297a0 ', '#eaab00 ', '#00b588 ', '#0098db',
                  '#f8a98e ', '#012d56']
    
    ### 지도시각화
    for idx in range(len(dong_list)):
        layer = folium.FeatureGroup(name="{} : {}".format(idx+1,dong_list[idx]))
        
        color_style = color_list[idx]
        
        df_dong = df_map[df_map["dong"]==dong_list[idx]].reset_index(drop=True)
        
        for val in df_dong["restaurant_name"]:
            i = df_dong[df_dong["restaurant_name"]==val].index
            
            lat = df_dong.iloc[i, 5].values
            lng = df_dong.iloc[i, 6].values
            name = df_dong.iloc[i, 2].values[0]
            marker_weight = 5
    
            folium.CircleMarker(location=[lat, lng],
                                fill_color=color_style,
                                fill_opacity=1,
                                color=color_style,
                                weight=marker_weight,
                                radius=marker_weight,
                                popup=name).add_to(layer)
            
        layer.add_to(rv_map)
    layer_control.add_to(rv_map)
    
    return rv_map
```

```python
### 소재지(동)/음식 종류/키워드 입력 시 식당을 추천해주는 함수
# - 기준: 키워드 수 200개 이상인 식당
def getRecommendRest(dong, category, category2):

    ## 입력한 동(dong)과 음식 종류(category)에 해당하는 식당명 리스트 생성
    list_dong_cat = df_cat_dong[(df_cat_dong["dong"]==dong)&(df_cat_dong["category"]==category)==True]["restaurant_name"]
    ## list_dong_cat 의 식당에 대한 식당명/평가 기준/평가 수/정규화된 평가 수 출력
    filtered_df_cat_count = df_cat_count[df_cat_count["restaurant_name"].isin(list_dong_cat)]
    ## 평가 수 200개 이상인 식당 중, 원하는 평가 기준(category2)에 부합하는 식당의 상위 5건 필터링
    df_recommended_rest = filtered_df_cat_count[(filtered_df_cat_count["naver_cat_review"]==category2)\
                                              & (filtered_df_cat_count['naver_cat_count'] > 0)].head(5).sort_values("normalized", ascending=False)
    
    recommend_rest = df_recommended_rest["restaurant_name"].to_list()
    if len(recommend_rest) != 0:
        print("당신이 원하는 기준에 해당하는 식당을 최대 5개까지 추천드려요!\n", recommend_rest)
    elif len(recommend_rest) == 0:
        response = "죄송합니다. 해당 기준에 부합하는 식당이 없습니다."
        return response
```

#### 상호명 별 특성 Top5 시각화 함수

```python
### 식당 특성 시각화 함수
def getRestPlot(res):
    import matplotlib.pyplot as plt
    import seaborn as sns
    plt.rc("font", family="Malgun Gothic")

    ## 시각화 할 상호명의 데이터만 필터링
    df_filtered = df_cat_count[(df_cat_count["restaurant_name"]==res)==True].head(5)

    ## 시각화
    plt.figure(figsize=(7, 3))
    ax = sns.barplot(x=df_filtered["naver_cat_review"], y=df_filtered["naver_cat_count"], palette="pastel")
    plt.title("{}에 대한 키워드 Top5".format(res))
    plt.xlabel("Keywords")
    plt.ylabel("count")
    ## y축 범위설정: 여유를 두었습니다....
    plt.ylim(df_filtered["naver_cat_count"].min() - df_filtered["naver_cat_count"].min()/10,
             df_filtered["naver_cat_count"].max() + df_filtered["naver_cat_count"].max()/10)
    # 각 막대 상단에 값을 표시
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width() / 2., p.get_height(), f"{p.get_height():.0f}", 
                fontsize=12, color='black', ha='center', va='bottom')
    return plt.show()
```

```python
"""
 위 파일 최상단에서 정의한 review_loc의 데이터를 가지고 지도 시각화를 진행
"""

### 지도시각화 함수
def getMap(store_list):
    import folium
    """
    store_list = ["동 가게이름", "동 가게이름"]으로 정의
    """
    ### 지도 시작 위치 지정
    lat = float(df_map[df_map["restaurant_name"]==store_list[0]]["위도"].iloc[0])
    lng = float(df_map[df_map["restaurant_name"]==store_list[0]]["경도"].iloc[0])
    
    ### 지도시각화 정의
    store_map = folium.Map(location=[lat, lng], tiles="CartoDB positron", zoom_start=30)
    
    for val in store_list:
        idx = df_map[df_map["restaurant_name"]==val].index

        lat = df_map.iloc[idx, 5].values
        lng = df_map.iloc[idx, 6].values
        name = df_map.iloc[idx, 2].values[0]

        custom_icon = folium.CustomIcon(icon_image="./img/밥.png", icon_size=(50, 50))
        marker2 = folium.Marker(location=[lat, lng], icon=custom_icon, popup=name)
        marker2.add_to(store_map)

    
    return store_map
```
