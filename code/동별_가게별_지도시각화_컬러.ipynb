```python
import pandas as pd
import folium
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic")
```

# 파일 불러들이기

```python
df_map = pd.read_excel("./data/review_latlon_uniq.xlsx")
df_map
```

# 동별 가게 시각화

```python
### 지도시각화 정의
rv_map = folium.Map(location=[35.1665, 129.0431], tiles="CartoDB positron", zoom_start=13)
layer_control = folium.LayerControl()

### 값 필터 설정
# - 동 리스트 고윳값 정의
dong_list = df_map["dong"].unique()
# - 동별 마커 색상 지정
colors = {"양정동":"#FFC0CB",  # 밝은 분홍색
          "개금동":"#FFD700",  # 황금색
          "가야동":"#228B22",  # 숲색
          "당감동":"#800080",  # 자주색
          "전포동":"#FFA500",  # 향긋한 오렌지
          "연지동":"#87CEEB",  # 하늘색
          "범전동":"#00FF00",  # 연두색
          "초읍동":"#FF6F61",  # 코랄 핑크
          "범천동":"#808000",  # 올리브 그린
          "부전동":"#9370DB",  # 조화로운 자색
          "부암동":"#1E90FF"}   # 바다색 블루

### 지도시각화
for idx in range(len(dong_list)):
    layer = folium.FeatureGroup(name="{} : {}".format(idx+1,dong_list[idx]))
    
    color_style = colors[dong_list[idx]]
    
    df_dong = df_map[df_map["dong"]==dong_list[idx]].reset_index(drop=True)
    
    for val in df_dong["restaurant_name"]:
        i = df_dong[df_dong["restaurant_name"]==val].index
        
        lat = df_dong.iloc[i, 5].values
        lng = df_dong.iloc[i, 6].values
        name = df_dong.iloc[i, 2].values[0]
        marker_weight = 5

        folium.CircleMarker(location=[lat, lng],
                            fill_color=color_style,
                            fill_opacity=1,
                            color=color_style,
                            weight=marker_weight,
                            radius=marker_weight,
                            tooltip=name).add_to(layer)
        
    layer.add_to(rv_map)
layer_control.add_to(rv_map)    
rv_map
```

### 리스트로 가게 입력 시 지도를 반환하는 함수

```python
"""
 위 파일 최상단에서 정의한 review_latlon_uniq.xlsx 데이터를 가지고 지도 시각화를 진행
"""
### 동별 색상 지정
colors = {"양정동":"#FFC0CB",  # 밝은 분홍색
          "개금동":"#00FFBF",  # 민트 그린
          "가야동":"#FFD700",  # 황금색
          "당감동":"#800080",  # 자주색
          "전포동":"#FFA500",  # 향긋한 오렌지
          "연지동":"#87CEEB",  # 하늘색
          "범전동":"#00FF00",  # 연두색
          "초읍동":"#FF6F61",  # 코랄 핑크
          "범천동":"#808000",  # 올리브 그린
          "부전동":"#9370DB",  # 조화로운 자색
          "부암동":"#1E90FF"}   # 바다색 블루


### 지도시각화 함수
def getMap(store_list):
    """
    store_list = ["동 가게이름", "동 가게이름"]으로 정의
    """
    
    ### 지도시각화 정의
    store_map = folium.Map(location=[35.1665, 129.0431], tiles="CartoDB positron", zoom_start=13)
    
    for val in store_list:
        idx = df_map[df_map["restaurant_name"]==val].index

        lat = df_map.iloc[idx, 5].values
        lng = df_map.iloc[idx, 6].values
        name = df_map.iloc[idx, 2].values[0]
        marker_weight = 5
        color_dong = colors[val[:3]]
        
        folium.CircleMarker(location=[lat, lng],
                            fill_color=color_dong,
                            fill_opacity=1,
                            color=color_dong,
                            weight=marker_weight,
                            radius=marker_weight,
                            tooltip=name).add_to(store_map)
    
    return store_map
```

### 키워드별 가게 시각화

```python
cat_loc = pd.read_excel("./data/cat_latlon.xlsx")
```

```python
############### 추출하고 싶은 키워드 정의하기
keyword = "음식이 맛있어요"
#################################################



### 리뷰 키워드 한 행에 합치기
# - {가게명:[키워드종합]} 딕셔너리로 정의
kw_dict = dict(cat_loc.groupby("restaurant_name")["naver_cat_review"].unique().items())

# - cat_uniq에 가게명 고윳값과 키워드 종합열 추가
cat_uniq = cat_loc.groupby("restaurant_name").head(1).reset_index(drop=True)
cat_uniq["total_review"] = [kw_dict[cat_uniq.loc[i,"restaurant_name"]] \
                            for i in range(len(cat_uniq))]


### "혼밥하기 좋아요"인 가게들만 뽑아내기
str_list = [cat_uniq["restaurant_name"][i] for i in range(len(cat_uniq)) \
            if (keyword in cat_uniq["total_review"].apply(lambda x: x[:5]).tolist()[i])]

print(str_list)
```

```python
### 지도시각화 함수
def getMap(store_list):
    """
    store_list = ["동 가게이름", "동 가게이름"]으로 정의
    """

    ### 동별 색상 지정
    colors = {"양정동":"#FFC0CB",  # 밝은 분홍색
              "개금동":"#00FFBF",  # 민트 그린
              "가야동":"#FFD700",  # 황금색
              "당감동":"#800080",  # 자주색
              "전포동":"#FFA500",  # 향긋한 오렌지
              "연지동":"#87CEEB",  # 하늘색
              "범전동":"#00FF00",  # 연두색
              "초읍동":"#FF6F61",  # 코랄 핑크
              "범천동":"#808000",  # 올리브 그린
              "부전동":"#9370DB",  # 조화로운 자색
              "부암동":"#1E90FF"}   # 바다색 블루
    
    ### 지도시각화 정의
    store_map = folium.Map(location=[35.1665, 129.0431], tiles="CartoDB positron", zoom_start=13)
    
    for val in store_list:
        idx = cat_uniq[cat_uniq["restaurant_name"]==val].index

        lat = cat_uniq.iloc[idx, 7].values
        lng = cat_uniq.iloc[idx, 8].values
        name = cat_uniq.iloc[idx, 5].values[0]
        marker_weight = 5
        color_dong = colors[val[:3]]

        folium.CircleMarker(location=[lat, lng],
                            fill_color=color_dong,
                            fill_opacity=1,
                            color=color_dong,
                            weight=marker_weight,
                            radius=marker_weight,
                            tooltip=name).add_to(store_map)
    
    return store_map
```
